```python
df = pd.read_csv('mozilla-metrics.csv')
df = df.drop(['file', 'crashes'], axis=1)
df
```

```python
arff.dump('mozilla-metrics.arff'
      , df.values
      , relation='mozilla'
      , names=df.columns)
```
